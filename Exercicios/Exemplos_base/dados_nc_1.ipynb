# Trabalho com dados cientificos `netCDF`

network Commom Data Form (netCDF), é um formato de arquivo de armazenamento de dados multidimensional. Utilizado principalmente pela comunidade científica. Armazena variáveis como:
* Umidade
* Pressão
* Velocidade do vento
* Direção do vento
* etc.

Principais características do formato:
- **Auto descritiva**, o arquivo contem informação sob o conteudo.
- **Potavel**, disponivel para diferentes formas de armazenamento (string, integers, ponto flutuante).
- **Escalavel**, interfaces de netcdf podem acessar facilmente datasets em diferentes formatos.
- **Anexavel**, podem anexar novos conjuntos de dados sem modificar os existentes.
- **Compartilhavel** (Shareable), multiples usuarios podem acessar o mesmo arquivo.
- **Arquivável**, versões mais antigas tem suporte por versões atuais e futuras.

Focaremos na parte da velocidade do vento

```python
# topografia
print('Topografia:\n',topo[10, 5:10, :5])

#velocidade
print('Velocidade em u:\n', vel_u[10, 5:10, :5])

print(vel_u.shape)
```

```python
# obetr a velocidade do vento num determinado ponto e plota ele
import numpy as np

vel_temp = np.sqrt(np.power(vel_u[:, 60, 60],2) + np.power(vel_v[0, 60, 60],2))
print(vel_temp)
```

```python
#pot da velocidade
from matplotlib import pyplot as plt
plt.plot(vel_temp, '--')
plt.xlabel('Tempo')
plt.ylabel('Velocidade (m/s)')
plt.title('Velocidade do vento para o centro do dominio')
```

```python
# pirmeiro tempo, a superficie completa
vel_sup = np.sqrt(np.power(vel_u[0, :,:],2) + np.power(vel_v[0, :,:],2))

# cuidado!!!
# deixando a figura de cabeça para baixo
plt.imshow(vel_sup)
```

```python
lats = nc.variables['XLAT'][0,:,:]
lons = nc.variables['XLONG'][0,:,:]
plt.pcolormesh(lons, lats, vel_sup, cmap='jet')
plt.colorbar()
```
